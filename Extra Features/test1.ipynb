```python
df = pd.read_csv("Dataset/IndianFoodDatasetCSV.csv")
```

```python
df = df.drop(['Srno', 'RecipeName','Ingredients', 'Instructions' ],axis=1)
```

```python
df.rename(columns = {'TranslatedRecipeName':'RecipieName','TranslatedIngredients':'Ingredients','TranslatedInstructions':'Instructions'}, inplace = True)
```

```python
translator = Translator()
```

```python
translated_recipes = []
i = 0
for recipe in df['RecipieName']:
    lang = detect(recipe)
    if lang == 'hi':
        translated_recipe = translator.translate(recipe, src='hi', dest='en')
        df['RecipieName'][i]=translated_recipe.text
        i=i+1
    else:
        df['RecipieName'][i]= recipe
        i=i+1
```

```python
translated_instuctions = []
i = 0
list=[]
for recipe in df['Instructions']:
    lang = detect(recipe)
    try:
        if lang == 'hi':
            list.append(i)
    except:
        pass
    i=i+1
    
```

```python
df.drop(df.index[list],inplace=True)
```
